```python
from pathlib import Path

from canvod.readers import Rnxv3Obs
```

```python
# Example file path (adjust to your actual data location)
rinex_file = Path(
    "/Users/work/Developer/GNSS/canvodpy/packages/canvod-readers/tests/test_data/01_Rosalia/01_reference/01_GNSS/01_raw/25001/rref001a00.25o"
)

# Check if file exists
if rinex_file.exists():
    # Create reader instance by passing the file path
    reader = Rnxv3Obs(fpath=rinex_file)
    print(f"âœ“ Successfully loaded: {rinex_file.name}")
    print(f"âœ“ RINEX version: {reader.header.version}")
    print(f"âœ“ Marker name: {reader.header.marker_name}")
    print(f"âœ“ Systems: {reader.header.systems}")
else:
    print(f"âœ— File not found: {rinex_file}")
    reader = None
```

```python
ds = reader.to_ds(
    keep_rnx_data_vars=[
        "SNR",
        "Doppler",
        "Phase",
        "Pseudorange",
    ],  # Keep only SNR observations
    write_global_attrs=True,  # Include comprehensive metadata
)
ds
```
